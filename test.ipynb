```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Filter the dataset
patient_p001_details = df[
    (df['patient_id'] == 'P001') & 
    (df['admission_id'] == 'ADM002') & 
    (df['test_code'] == 'HEMOGLOBIN')
].copy()

# 2. Ensure numerical and datetime data types are correct
patient_p001_details['collection_datetime'] = pd.to_datetime(patient_p001_details['collection_datetime'])
patient_p001_details['value_numeric'] = pd.to_numeric(patient_p001_details['value_numeric'])
patient_p001_details['reference_low'] = pd.to_numeric(patient_p001_details['reference_low'])
patient_p001_details['reference_high'] = pd.to_numeric(patient_p001_details['reference_high'])

# Sort by time
patient_p001_details = patient_p001_details.sort_values('collection_datetime')

# 3. Separate Normal vs. Abnormal results
is_normal = patient_p001_details['abnormal_flag'].str.lower().str.strip() == 'normal'
normal_points = patient_p001_details[is_normal]
abnormal_points = patient_p001_details[~is_normal]

# Get the reference values for plotting boundaries (safely handling missing values if any)
try:
    ref_low = patient_p001_details['reference_low'].dropna().iloc[0]
    ref_high = patient_p001_details['reference_high'].dropna().iloc[0]
except IndexError:
    ref_low, ref_high = None, None

# 4. Plotting
plt.figure(figsize=(12, 6))

# A. Shade the normal reference range area if reference values exist
if ref_low is not None and ref_high is not None:
    plt.fill_between(
        patient_p001_details['collection_datetime'],
        patient_p001_details['reference_low'],
        patient_p001_details['reference_high'],
        color='green',
        alpha=0.1,
        label=f'Normal Range ({ref_low} - {ref_high} g/dL)'
    )
    # Draw reference boundary lines
    plt.axhline(y=ref_low, color='red', linestyle='--', alpha=0.4, label=f'Ref Low ({ref_low})')
    plt.axhline(y=ref_high, color='red', linestyle='--', alpha=0.4, label=f'Ref High ({ref_high})')

# B. Draw the progress line connecting all points
plt.plot(
    patient_p001_details['collection_datetime'], 
    patient_p001_details['value_numeric'], 
    color='#5c768d', 
    linestyle='-', 
    linewidth=2,
    alpha=0.8,
    zorder=1
)

# C. Plot normal points as green dots
plt.scatter(
    normal_points['collection_datetime'], 
    normal_points['value_numeric'], 
    color='green', 
    s=80, 
    zorder=2, 
    label='Normal Result'
)

# D. Plot abnormal points as prominent red X markers
plt.scatter(
    abnormal_points['collection_datetime'], 
    abnormal_points['value_numeric'], 
    color='red', 
    marker='X', 
    s=120, 
    zorder=3, 
    label='Abnormal Result'
)

# E. Annotate abnormal points on the graph (displays e.g. "High (18.5)")
for idx, row in abnormal_points.iterrows():
    plt.annotate(
        f"{row['abnormal_flag']} ({row['value_numeric']})", 
        (row['collection_datetime'], row['value_numeric']),
        textcoords="offset points", 
        xytext=(0, 10), 
        ha='center', 
        color='red',
        weight='bold',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="red", lw=0.5, alpha=0.8)
    )

# Customize the chart labels and layout
plt.title('Hemoglobin Levels Progress (Patient: P001 | Admission: ADM002)', fontsize=14, pad=15)
plt.xlabel('Collection Date & Time', fontsize=12)
plt.ylabel('Hemoglobin Value (g/dL)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.xticks(rotation=30, ha='right')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Define the tests we want to plot
test_codes = ['HEMOGLOBIN', 'WBC', 'RBC']

# 2. Filter the dataset for Patient P001, Admission ADM002, and the selected tests
patient_p001_details = df[
    (df['patient_id'] == 'P001') & 
    (df['admission_id'] == 'ADM002') & 
    (df['test_code'].isin(test_codes))
].copy()

# 3. Ensure numerical and datetime data types are correct
patient_p001_details['collection_datetime'] = pd.to_datetime(patient_p001_details['collection_datetime'])
patient_p001_details['value_numeric'] = pd.to_numeric(patient_p001_details['value_numeric'])
patient_p001_details['reference_low'] = pd.to_numeric(patient_p001_details['reference_low'])
patient_p001_details['reference_high'] = pd.to_numeric(patient_p001_details['reference_high'])

# Sort chronologically
patient_p001_details = patient_p001_details.sort_values('collection_datetime')

# 4. Create subplots (one subplot per test, sharing the same X-axis timeline)
fig, axes = plt.subplots(len(test_codes), 1, figsize=(12, 4 * len(test_codes)), sharex=True)

# If only one test is defined, wrap it in a list to keep the loop iterable
if len(test_codes) == 1:
    axes = [axes]

for ax, test_code in zip(axes, test_codes):
    # Filter data for this specific test
    test_df = patient_p001_details[patient_p001_details['test_code'] == test_code]
    
    if test_df.empty:
        ax.text(0.5, 0.5, f"No data found for {test_code}", ha='center', va='center', fontsize=12)
        ax.set_title(f"{test_code} Progress")
        continue

    # Get measurement unit dynamically
    unit = test_df['unit'].dropna().iloc[0] if not test_df['unit'].dropna().empty else ''
    
    # Separate Normal vs. Abnormal results
    is_normal = test_df['abnormal_flag'].str.lower().str.strip() == 'normal'
    normal_points = test_df[is_normal]
    abnormal_points = test_df[~is_normal]
    
    # Get reference boundaries safely
    try:
        ref_low = test_df['reference_low'].dropna().iloc[0]
        ref_high = test_df['reference_high'].dropna().iloc[0]
    except IndexError:
        ref_low, ref_high = None, None

    # A. Shade the normal reference range if values are available
    if ref_low is not None and ref_high is not None:
        ax.fill_between(
            test_df['collection_datetime'],
            test_df['reference_low'],
            test_df['reference_high'],
            color='green',
            alpha=0.08,
            label=f'Normal Range ({ref_low} - {ref_high} {unit})'
        )
        # Reference limit lines
        ax.axhline(y=ref_low, color='red', linestyle='--', alpha=0.3, label=f'Ref Low ({ref_low})')
        ax.axhline(y=ref_high, color='red', linestyle='--', alpha=0.3, label=f'Ref High ({ref_high})')

    # B. Plot the trend line
    ax.plot(
        test_df['collection_datetime'], 
        test_df['value_numeric'], 
        color='#5c768d', 
        linestyle='-', 
        linewidth=2,
        alpha=0.8,
        zorder=1
    )

    # C. Plot normal points (green dots)
    ax.scatter(
        normal_points['collection_datetime'], 
        normal_points['value_numeric'], 
        color='green', 
        s=70, 
        zorder=2, 
        label='Normal Result'
    )

    # D. Plot abnormal points (red X markers)
    ax.scatter(
        abnormal_points['collection_datetime'], 
        abnormal_points['value_numeric'], 
        color='red', 
        marker='X', 
        s=100, 
        zorder=3, 
        label='Abnormal Result'
    )

    # E. Annotate abnormal points (e.g. "High (15.6)")
    for idx, row in abnormal_points.iterrows():
        ax.annotate(
            f"{row['abnormal_flag']} ({row['value_numeric']})", 
            (row['collection_datetime'], row['value_numeric']),
            textcoords="offset points", 
            xytext=(0, 10), 
            ha='center', 
            color='red',
            weight='bold',
            fontsize=8,
            bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="red", lw=0.5, alpha=0.8)
        )

    # Chart formatting for this specific subplot
    ax.set_title(f'{test_code} Progress over Time', fontsize=12, pad=10, weight='bold')
    ax.set_ylabel(f'Value ({unit})', fontsize=10)
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend(loc='upper left', fontsize=9)

# Format the shared X-axis on the bottom plot
plt.xlabel('Collection Date & Time', fontsize=12)
plt.xticks(rotation=30, ha='right')

# Overall Figure Title
plt.suptitle('Patient P001 Lab Progress Summary (Admission: ADM002)', fontsize=15, weight='bold', y=0.99)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# 1. Filter the dataset for Patient P001 and Admission ADM002
patient_p001_details = df[
    (df['patient_id'] == 'P001') & 
    (df['admission_id'] == 'ADM002')
].copy()

# 2. Ensure numerical and datetime data types are correct
patient_p001_details['collection_datetime'] = pd.to_datetime(patient_p001_details['collection_datetime'])
patient_p001_details['value_numeric'] = pd.to_numeric(patient_p001_details['value_numeric'], errors='coerce')
patient_p001_details['reference_low'] = pd.to_numeric(patient_p001_details['reference_low'], errors='coerce')
patient_p001_details['reference_high'] = pd.to_numeric(patient_p001_details['reference_high'], errors='coerce')

# Sort chronologically
patient_p001_details = patient_p001_details.sort_values('collection_datetime')

# 3. Dynamically get all unique test codes present for this patient admission (sorted alphabetically)
test_codes = sorted(patient_p001_details['test_code'].dropna().unique())

# 4. Set up a dynamic grid layout (2 columns)
n_cols = 2
n_rows = math.ceil(len(test_codes) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 4 * n_rows))

# Flatten axes for easy 1D iteration
if len(test_codes) == 1:
    axes = [axes]
else:
    axes = axes.flatten()

# 5. Plot each test in its own subplot
for i, test_code in enumerate(test_codes):
    ax = axes[i]
    
    # Filter data for this specific test in the loop
    test_df = patient_p001_details[patient_p001_details['test_code'] == test_code]
    
    # Skip if no numeric data exists
    if test_df['value_numeric'].dropna().empty:
        ax.text(0.5, 0.5, f"No numeric data for {test_code}", ha='center', va='center', fontsize=12)
        ax.set_title(f"{test_code} Progress")
        continue

    # Get measurement unit dynamically
    unit = test_df['unit'].dropna().iloc[0] if not test_df['unit'].dropna().empty else ''
    
    # Separate Normal vs. Abnormal results
    is_normal = test_df['abnormal_flag'].str.lower().str.strip() == 'normal'
    normal_points = test_df[is_normal]
    abnormal_points = test_df[~is_normal]
    
    # Get reference boundaries safely
    try:
        ref_low = test_df['reference_low'].dropna().iloc[0]
        ref_high = test_df['reference_high'].dropna().iloc[0]
    except IndexError:
        ref_low, ref_high = None, None

    # A. Shade the normal reference range if values are available
    if ref_low is not None and ref_high is not None:
        ax.fill_between(
            test_df['collection_datetime'],
            test_df['reference_low'],
            test_df['reference_high'],
            color='green',
            alpha=0.08,
            label=f'Normal Range ({ref_low} - {ref_high} {unit})'
        )
        # Reference limit lines
        ax.axhline(y=ref_low, color='red', linestyle='--', alpha=0.3, label=f'Ref Low ({ref_low})')
        ax.axhline(y=ref_high, color='red', linestyle='--', alpha=0.3, label=f'Ref High ({ref_high})')

    # B. Plot the trend line
    ax.plot(
        test_df['collection_datetime'], 
        test_df['value_numeric'], 
        color='#5c768d', 
        linestyle='-', 
        linewidth=2,
        alpha=0.8,
        zorder=1
    )

    # C. Plot normal points (green dots)
    ax.scatter(
        normal_points['collection_datetime'], 
        normal_points['value_numeric'], 
        color='green', 
        s=70, 
        zorder=2, 
        label='Normal Result'
    )

    # D. Plot abnormal points (red X markers)
    ax.scatter(
        abnormal_points['collection_datetime'], 
        abnormal_points['value_numeric'], 
        color='red', 
        marker='X', 
        s=100, 
        zorder=3, 
        label='Abnormal Result'
    )

    # E. Annotate abnormal points
    for idx, row in abnormal_points.iterrows():
        ax.annotate(
            f"{row['abnormal_flag']} ({row['value_numeric']})", 
            (row['collection_datetime'], row['value_numeric']),
            textcoords="offset points", 
            xytext=(0, 10), 
            ha='center', 
            color='red',
            weight='bold',
            fontsize=8,
            bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="red", lw=0.5, alpha=0.8)
        )

    # Chart formatting for this specific subplot
    ax.set_title(f'{test_code} Progress', fontsize=12, pad=10, weight='bold')
    ax.set_ylabel(f'Value ({unit})', fontsize=10)
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend(loc='upper left', fontsize=9)
    ax.tick_params(axis='x', rotation=30)

# Hide any unused subplots in the grid
for j in range(len(test_codes), len(axes)):
    fig.delaxes(axes[j])

# Overall layout details
plt.suptitle('Patient P001 Lab Progress Summary (Admission: ADM002) - All Tests', fontsize=16, weight='bold', y=0.99)
plt.tight_layout()
plt.show()
```
